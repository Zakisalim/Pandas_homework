```python
nine_grade_math = student_data.loc[student_data["grade"] == "9th"].groupby("school_name")["math_score"].mean()
ten_grade_math = student_data.loc[student_data["grade"] == "10th"].groupby("school_name")["math_score"].mean()
eleven_grade_math = student_data.loc[student_data["grade"] == "11th"].groupby("school_name")["math_score"].mean()
twelfe_grade_math = student_data.loc[student_data["grade"] == "12th"].groupby("school_name")["math_score"].mean()

math_by_grade = pd.DataFrame(list(zip(nine_grade_math, ten_grade_math, eleven_grade_math, twelfe_grade_math)), 
               columns =["9th", "10th", "11th", "12th"])

schools= school_data.set_index("School ID")["school_name"] 
math_by_grade["School Name"]=schools
math_by_grade=math_by_grade[["School Name", "9th", "10th", "11th", "12th"]]
math_by_grade.reset_index(drop=True)

math_by_grade
```

```python
nine_grade_reading = student_data.loc[student_data["grade"] == "9th"].groupby("school_name")["reading_score"].mean()
ten_grade_reading = student_data.loc[student_data["grade"] == "10th"].groupby("school_name")["reading_score"].mean()
eleven_grade_reading = student_data.loc[student_data["grade"] == "11th"].groupby("school_name")["reading_score"].mean()
twelfe_grade_reading = student_data.loc[student_data["grade"] == "12th"].groupby("school_name")["reading_score"].mean()

reading_by_grade = pd.DataFrame(list(zip(nine_grade_reading, ten_grade_reading, eleven_grade_reading, twelfe_grade_reading)), 
               columns =["9th", "10th", "11th", "12th"])

schools= school_data.set_index("School ID")["school_name"] 
reading_by_grade["School Name"]=schools
reading_by_grade= reading_by_grade[["School Name", "9th", "10th", "11th", "12th"]]
reading_by_grade.reset_index(drop=True)

reading_by_grade
```

```python
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
school_summary["spending"] = pd.cut(school_summary["Budget Per Student"], spending_bins, labels=group_names)
school_by_spending= school_summary.groupby("spending")
Average_math= school_by_spending["Average Math Score"].mean()
Average_reading= school_by_spending["Average Reading Score"].mean()
pass_math = school_summary[school_summary["Average Math Score"] >= 70].groupby("spending")["Budget"].count()/school_by_spending["Budget"].count()
pass_reading = school_summary[school_summary["Average Reading Score"] >= 70].groupby("spending")["Budget"].count()/school_by_spending["Budget"].count()
schools_by_spending= pd.DataFrame(list(zip(Average_math, Average_reading, pass_math, pass_reading)), 
                                     columns= ["Average Math", "Average Reading", "% Passing Math", "% Passing Reading"])
schools_by_spending
```
